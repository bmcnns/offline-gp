```python
import os
import pickle
from pyoperon.sklearn import SymbolicRegressor
from gp.model import ImitationLearner
import matplotlib.pyplot as plt
import gymnasium as gym
import numpy as np
from IPython.display import display, Video
import os
```

```python
complete_models = list(filter(lambda x: not x.split('.')[0][-1].isdigit(), os.listdir("models/gp/")))
```

```python
models = []
for model_file in complete_models:
    with open(f"models/gp/{model_file}", 'rb') as f:
        print(f"Loading model {model_file}...")
        model = pickle.load(f)
        models.append(model)
```

```python
hopper = models[0]
```

```python
for i, env_name in enumerate(['hopper']):
    action_ids = models[i].regressors.keys()
    
    for _, regressor in models[i].regressors.items():
        mdls = []
        mses = []
        res = [(s['objective_values'], s['tree'], s['minimum_description_length']) for s in regressor.pareto_front_]
        
        for obj, expr, mdl in res:
            print(obj, expr, regressor.get_model_string(expr, 3))
            mdls.append(mdl)
            mses.append(obj[0])

        print(mdls)
        print(mses)
        
    plt.bar(x=action_ids, height=mdls)
    plt.title(f"{env_name}: Program Length of Action Selectors")
    plt.show()
    
    plt.bar(x=action_ids, height=mses)
    plt.title(f"{env_name}: Mean-Squared Error of Actions")
    plt.show()
        
```

```python
model = hopper.regressors['action0']
s = model.get_model_string(model.model_, 3)
print(s)
```

```python
def run_model(env_name, model):
    env = gym.make(env_name, render_mode='rgb_array')
    env = gym.wrappers.RecordVideo(env, "videos", episode_trigger=lambda x: True)
    
    obs = env.reset()[0]

    action = np.zeros(env.action_space.shape[0])
    print(action.shape)

    terminated, truncated = False, False
    
    while not (terminated or truncated):
        # Get the next action
        for i, action_selector in enumerate(model.regressors.items()):
            _, action_selector = action_selector

            scaler = model.scalers.get(f"action{i}")
            scaled_obs = scaler.transform(obs.reshape(1,-1))
            action[i] = action_selector.predict(scaled_obs).item()
            
        action = np.clip(action, a_min=0.0, a_max=1.0)
        print(action)
        obs, reward, terminated, truncated, _ = env.step(action)

    env.close()
```

```python
run_model("Hopper-v5", hopper)
```
