```python
from S_run_aifeynman import run_aifeynman

def train(dataset_name, num_independent_runs, save_dir):
    X, y = load_dataset(dataset_name)
    
    for run in range(num_independent_runs):
        dataset_path = f"{save_dir}/run_{run}/"
        os.makedirs(dataset_path, exist_ok=True)
    
        for action in range(y.shape[1]):
            with open(f'{dataset_path}/action{action+1}.txt', 'w') as f:
                for X_i, y_i in zip(X, y[:, action]):
                    f.write(f'{" ".join(map(str, X_i))} {y_i}\n')
    
    for run in range(num_independent_runs):
        dataset_path = f"{save_dir}/run_{run}/"
        for action in range(y.shape[1]):
            try:
                run_aifeynman(dataset_path,f"action{action+1}.txt",3,"14ops.txt")
            except:
                continue
```

```python
train("Minimal-Hopper-Expert-v5", num_independent_runs=1, save_dir="models/imitation_learners/aifeynman")
```
